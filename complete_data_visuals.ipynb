```python
#Pandas for data manipulation (pd is shortcut name)
#numpy for computaion
#matplotlib for plotting
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

# loading all 4 data files

```python
#pd.read_csv reads csv files and return a data Frame, names=['X','Y'] sets the colomn names
s1=pd.read_csv('Meditation experiment with tagging-1/NFB_17-4-2019_20-48._FP2csv.csv',names=['X','Y'])
s2=pd.read_csv('Meditation experiment with tagging-1/NFB_17-4-2019_22-35_F4.csv',names=['X','Y'])
s3=pd.read_csv('Meditation experiment with tagging-1/NFB_18-4-2019_0-13_O2.csv',names=['X','Y'])
s4=pd.read_csv('Meditation experiment with tagging-1/NFB_18-4-2019_21-17._P4. csv.csv',names=['X','Y'])
```

```python
#slicing the files as start contained some string values
s1=s1[3:]
s2=s2[3:]
s3=s3[12:]
s4=s4[3:]
```

```python
#converting the string types to floats ('1.23' to 1.23)
s1.X=s1.X.astype(float)
#reset the index values to start from 0
s1=s1.reset_index().values
```

# Value, tag and colour
0   unlabled    Black          
32  peak        red         
37  som med     blue        
38  good med    yellow          
39  mind wonder green                    
40  sleepy      white         

Mind Wonder and aleepy are 5-10 sec after it actually occoured


```python
#create a dictionary for all the labels
mapping= {0: ("black", "."), 37: ("blue", "."),39: ("green", "."), 40: ("white", "."),38: ("yellow", "."), 32: ("red", ".")}
```

# Fp2


```python
#loop through the file and find different y values, prot them one by one using the colour specified
#np.unique gives usique values of y
#plt.scatter plots a scatter plot
for c in np.unique(s1[:,2]):
    d = s1[s1[:,2] == c]
    plt.scatter(d[:,0], d[:,1], c=mapping[c][0], marker=mapping[c][1])
```
